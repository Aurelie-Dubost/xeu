```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Sample data generation
time = np.linspace(0, 2 * np.pi, 400)
data1 = np.sin(time ** 2)
data2 = np.cos(time ** 2)

# Define the figure and axes
fig1, ax1 = plt.subplots(figsize=(5, 3))
ax1.plot(time, data1)

fig2, ax2 = plt.subplots(figsize=(5, 3))
ax2.plot(time, data2)

# Display the figures in a tab layout for the first line
tab = widgets.Tab(children=[fig1.canvas, fig2.canvas])
tab.set_title(0, 'Tab 1')
tab.set_title(1, 'Tab 2')

# Creating a function to display two graphs per line
def create_graph_row(data1, data2):
    fig1, ax1 = plt.subplots(figsize=(5, 3))
    ax1.plot(time, data1)
    
    fig2, ax2 = plt.subplots(figsize=(5, 3))
    ax2.plot(time, data2)
    
    return widgets.HBox([fig1.canvas, fig2.canvas])

# Use VBox to stack the rows vertically
graph_rows = [create_graph_row(np.sin(time), np.cos(time)) for _ in range(4)]

# Combine the tab and graph rows in a final layout
final_layout = widgets.VBox([tab] + graph_rows)

# Display the final layout
display(final_layout)
```
