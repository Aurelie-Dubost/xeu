**Variables:**

### Functions

- add_custom_legend: Places a custom legend inside the plot.

### Axis and Helper Functions Chunk

```python
# Modified buffer logic to dynamically increase buffer ratio for small ranges
def calculate_data_range(values, buffer_ratio=0.1, min_buffer=1.0):
    """
    Calculate min/max with an enforced dynamic buffer to avoid tight y-axis limits.

    Parameters:
    - values: List or array of numerical values.
    - buffer_ratio: Fraction of the range to add as buffer.
    - min_buffer: Minimum buffer size.

    Returns:
    - vmin, vmax: Buffered y-axis limits.
    """
    vmin = np.min(values)
    vmax = np.max(values)
    range_ = vmax - vmin

    # Dynamically increase buffer for small ranges
    buffer = max(range_ * buffer_ratio, min_buffer)
    return vmin - buffer, vmax + buffer

def configure_spines_and_ticks(ax, show_top=False, show_right=False, 
                               x_ticks_position='bottom', y_ticks_position='left',
                               spine_color='black', spine_width=1.5):
    """
    Configure spines, tick positions, and their style.
    Always hides top and right spines if show_top=False and show_right=False.
    """
    # Set spine visibility
    ax.spines['top'].set_visible(show_top)
    ax.spines['right'].set_visible(show_right)

    # Configure spine appearance
    ax.spines['left'].set_color(spine_color)
    ax.spines['left'].set_linewidth(spine_width)
    ax.spines['bottom'].set_color(spine_color)
    ax.spines['bottom'].set_linewidth(spine_width)

    # Set tick positions
    ax.xaxis.set_ticks_position(x_ticks_position)
    ax.yaxis.set_ticks_position(y_ticks_position)

    # Configure tick parameters
    ax.tick_params(axis='x', colors='black', width=spine_width, direction='out')
    ax.tick_params(axis='y', colors='black', width=spine_width, direction='out')

def all_axis(ax, spine_color='black', spine_width=1.5):
    """
    Configure both x and y axis to have exterior tick marks, specific width, and color.
    Always end axis with plain tick marks and ensure smart white space usage.
    """
    configure_spines_and_ticks(ax, show_top=False, show_right=False, 
                               x_ticks_position='bottom', y_ticks_position='left',
                               spine_color=spine_color, spine_width=spine_width)
    ax.tick_params(axis='both', which='both', direction='out', length=5, width=spine_width)
    ax.margins(x=0.05, y=0.05)

def date_axis(ax, df, increment, date_format=None):
    """
    Configure the x-axis for date plotting based on the data length and increment (in months or years).

    - Always start tick labeling from the last date in the dataset.
    - The last month should always be displayed.
    - If increment < 12, use format '%b-%y' (e.g., 'Jan-23')
    - If increment >= 12, use format '%Y' (e.g., '2023')
    """
    # Ensure the index is datetime
    if not isinstance(df.index, pd.DatetimeIndex):
        df = df.copy()
        df.index = pd.to_datetime(df.index)

    dates = df.index
    last_date = dates.max()
    first_date = dates.min()

    # Determine date format if not provided
    if date_format is None:
        if increment < 12:
            date_format = '%b-%y'
        else:
            date_format = '%Y'

    ax.set_xlim(first_date, last_date)

    # Compute tick locations by stepping backward from last_date
    tick_dates = []
    current_date = last_date
    while current_date >= first_date:
        tick_dates.append(current_date)
        offset = pd.DateOffset(months=increment)
        current_date = current_date - offset

    tick_dates = sorted(tick_dates)
    ax.set_xticks(tick_dates)
    ax.set_xticklabels([d.strftime(date_format) for d in tick_dates], rotation=0, ha='center')

import math

def value_axis(ax, df, value_type='normal', decimals=2):
    """
    Configure the y-axis to minimize white space and use "nice" rounded ticks.

    Parameters:
    - ax: Matplotlib axis object
    - df: DataFrame or dict-like containing 'value'
    - value_type: 'normal' (default) or 'percentage'
    - decimals: Number of decimals for tick labels
    """
    # Extract values
    if hasattr(df, 'values'):
        values = df['value'].values
    else:
        values = np.array(list(df.values()))

    # Find the exact min and max
    vmin, vmax = np.min(values), np.max(values)

    # Function to determine the nearest "nice" step
    def nice_step(value_range, num_ticks=6):
        step = value_range / num_ticks
        magnitude = 10 ** math.floor(math.log10(step))  # Order of magnitude
        residual = step / magnitude
        if residual <= 1.5:
            nice = 1
        elif residual <= 3:
            nice = 2
        elif residual <= 7:
            nice = 5
        else:
            nice = 10
        return nice * magnitude

    # Compute step size and adjust limits
    range_ = vmax - vmin
    step = nice_step(range_)
    vmin = math.floor(vmin / step) * step
    vmax = math.ceil(vmax / step) * step

    # Generate ticks
    ticks = np.arange(vmin, vmax + step, step)
    ax.set_yticks(ticks)

    # Format tick labels
    if value_type == "percentage":
        formatted_ticks = [f'{tick:.{decimals}f}%' for tick in ticks]
    else:
        formatted_ticks = [f'{tick:.{decimals}f}' for tick in ticks]

    ax.set_yticklabels(formatted_ticks)
    ax.set_ylim(vmin, vmax)

def calculate_data_range(values, buffer_ratio=0.05, min_buffer=1.0):
    """
    Calculate the min and max values with a dynamic buffer.

    Parameters:
    - values: List or array of numerical values.
    - buffer_ratio: Proportion of the range to add as buffer.
    - min_buffer: Minimum buffer value for small ranges.

    Returns:
    - vmin: Buffered minimum value.
    - vmax: Buffered maximum value.
    """
    vmin, vmax = np.min(values), np.max(values)
    range_ = vmax - vmin
    buffer = max(range_ * buffer_ratio, min_buffer)
    return vmin - buffer, vmax + buffer



def value_axis(ax, series_dict, value_type='normal', decimals=2):
    """
    Configure the y-axis ticks to be rounded and well-spaced.

    Parameters:
    - ax: Matplotlib axis object.
    - series_dict: Dictionary of time series data.
    - value_type: 'normal' or 'percentage' for tick labels.
    - decimals: Decimal places for tick formatting.
    """
    vmin, vmax = calculate_combined_y_range(series_dict)
    range_ = vmax - vmin

    # Determine "nice" step size
    def nice_step(value_range, num_ticks=6):
        step = value_range / num_ticks
        magnitude = 10 ** math.floor(math.log10(step))
        residual = step / magnitude
        return magnitude * (1 if residual <= 1.5 else 2 if residual <= 3 else 5 if residual <= 7 else 10)

    step = nice_step(range_)
    ticks = np.arange(math.floor(vmin / step) * step, math.ceil(vmax / step) * step + step, step)

    # Format ticks
    formatted_ticks = [f'{tick:.{decimals}f}%' if value_type == 'percentage' else f'{tick:.{decimals}f}' for tick in ticks]
    ax.set_yticks(ticks)
    ax.set_yticklabels(formatted_ticks)
    ax.set_ylim(vmin, vmax)

def plot_last_value_horizontal_line(ax, df_or_dict, color='red', linestyle='--', linewidth=2):
    """
    Plot a horizontal line from the last value's y-coordinate to the y-axis.

    Parameters:
    - ax: matplotlib axis object to draw on
    - df_or_dict: data as a DataFrame (with a 'value' column) or a dict {time: value}
    - color: color of the line
    - linestyle: style of the line (default '--')
    - linewidth: width of the line
    """
    # Convert to DataFrame if needed
    if isinstance(df_or_dict, dict):
        data = pd.DataFrame(list(df_or_dict.items()), columns=['time', 'value']).set_index('time')
    else:
        data = df_or_dict.copy()

    # Ensure data is sorted
    data = data.sort_index()

    # Get the last point
    last_x = data.index[-1]
    last_y = data['value'].iloc[-1]

    # Plot a line from the last point to the y-axis
    ax.plot([last_x, ax.get_xlim()[0]], [last_y, last_y], color=color, linestyle=linestyle, linewidth=linewidth)

def calculate_combined_y_range(df_dict, buffer_ratio=0.05, min_buffer=1.0):
    """
    Calculate the combined y-axis range across all series with a buffer.

    Parameters:
    - df_dict: Dictionary of {series_name: DataFrame or dict}.
    - buffer_ratio: Proportion of the range to add as buffer.
    - min_buffer: Minimum buffer value to ensure spacing.

    Returns:
    - vmin: Minimum y value with buffer.
    - vmax: Maximum y value with buffer.
    """
    all_values = []
    for data in df_dict.values():
        if isinstance(data, dict):
            values = list(data.values())
        else:
            values = data['value'].values
        all_values.extend(values)
    
    # Calculate buffered range
    vmin, vmax = calculate_data_range(all_values, buffer_ratio, min_buffer)
    return vmin, vmax

def add_custom_legend(ax, buffer_ratio=0.025, series_dict=None):
    """
    Add a custom legend at a consistent position above the top y-axis limit, horizontally aligned.

    Parameters:
    - ax: Matplotlib axis object.
    - buffer_ratio: Proportion of the y-axis range to position the legend above the top.
    - series_dict: Dictionary of series to dynamically set legend columns.
    """
    # Calculate dynamic position just above the y-axis top
    ylim = ax.get_ylim()
    y_range = ylim[1] - ylim[0]
    y_legend = ylim[1] + (y_range * buffer_ratio)  # Add buffer to y-axis top

    # Dynamically calculate the number of legend columns
    ncol = len(series_dict.keys()) if series_dict else 1

    # Place the legend horizontally
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, y_legend / ylim[1]),
              ncol=ncol, frameon=False, handletextpad=1.0, columnspacing=1.5)
```

### Plot Function

```python
def simple_time_series_plot(series_dict, increment=10, value_type='normal', decimals=2,
                            spine_color='black', plot_last='yes', selected_last=None,
                            title=None, title_variable=None, buffer_ratio=0.05, min_buffer=1.0):
    """
    Plot multiple time series with enforced y-axis buffers, clean formatting, and a custom legend.
    """
    fig, ax = plt.subplots(figsize=(12, 6))
    colors = plt.cm.tab10.colors

    all_values = []  # Collect all values for y-axis range calculation
    for i, (series_name, data) in enumerate(series_dict.items()):
        ax.plot(data.index, data['value'], label=series_name, color=colors[i % len(colors)], lw=1.5)

        # Add horizontal lines for last values
        if plot_last.lower() == 'yes' or (selected_last and series_name in selected_last):
            last_value = data['value'].iloc[-1]
            ax.axhline(last_value, color=colors[i % len(colors)], linestyle='--', linewidth=1.5)

    # Calculate y-axis limits and apply with enforced buffer
    vmin = min(min(data['value']) for data in series_dict.values())
    vmax = max(max(data['value']) for data in series_dict.values())
    buffer = max((vmax - vmin) * buffer_ratio, min_buffer)

    vmin -= buffer
    vmax += buffer
    print(f"Global Y-Axis Limits: vmin={vmin}, vmax={vmax}, buffer={buffer}")

    ax.margins(0)  # Remove auto margins
    ax.set_ylim(vmin, vmax)  # Enforce Y-limits strictly

    # Configure axes appearance
    all_axis(ax, spine_color=spine_color)
    date_axis(ax, list(series_dict.values())[0], increment=increment)

    # Add custom legend
    add_custom_legend(ax, buffer_ratio=0.025, series_dict=series_dict)

    # Set title and labels
    if title:
        if title_variable:
            title = title.format(variable=title_variable)
        ax.set_title(title, pad=20)
    ax.set_xlabel("Time")
    ax.set_ylabel("Value")

    plt.show()
    return fig, ax
```
